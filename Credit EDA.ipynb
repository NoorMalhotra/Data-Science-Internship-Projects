```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")
```

```python
df = pd.read_csv("credit_risk_dataset.csv")
df.head()
```

```python
df.shape
df.info()
df.describe()
df.isnull().sum()
```

```python
sns.countplot(x='loan_status', data=df)
plt.title("Loan Status Distribution")
plt.show()
```

```python
sns.histplot(df['person_age'], bins=30, kde=True)
plt.title("Age Distribution")
plt.show()
```

```python
sns.histplot(df['person_income'], bins=30, kde=True)
plt.title("Income Distribution")
plt.show()
```

```python
sns.countplot(x='loan_intent', data=df)
plt.xticks(rotation=45)
plt.title("Loan Purpose")
plt.show()
```

```python
sns.countplot(x='person_home_ownership', data=df)
plt.title("Home Ownership")
plt.show()
```

```python
sns.boxplot(x='loan_status', y='person_income', data=df)
plt.title("Income vs Loan Status")
plt.show()
```

```python
sns.boxplot(x='loan_status', y='loan_amnt', data=df)
plt.title("Loan Amount vs Default")
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```

```python
sns.boxplot(x='loan_status', y='cb_person_cred_hist_length', data=df)
plt.title("Credit History vs Loan Status")
plt.show()
```

```python
sns.boxplot(x='loan_status', y='loan_int_rate', data=df)
plt.title("Interest Rate vs Loan Status")
plt.show()
```

# Observations:
# 1. Higher loan amounts tend to have higher default rates
# 2. Lower income individuals default more
# 3. Interest rates are higher for risky applicants
# 4. Credit history length impacts loan approval
